```python
X1 = pd.read_csv('72450003928.csv',sep=";")
X1.head()
```

```python
X2 = pd.read_csv('72546014933.csv',sep=";")
X2.head()
```

```python
X3 = pd.read_csv('72356013968.csv',sep=";")
X3.head()
```

```python
station_list = [X1.iloc[:100,:],X2.iloc[:100,:],X3.iloc[:100,:]]
station_data=pd.concat(station_list)
station_data
```

```python
station_data.shape
```

```python
plt.plot(station_data['DATE'],station_data['TMP'])
```

```python
station_data.groupby(["STATION","NAME"]).size()
```

```python
X1_date = station_data[station_data["STATION"]==72356013968]["DATE"]
X1_temp = station_data[station_data["STATION"]==72356013968]["TMP"]
X2_date = station_data[station_data["STATION"]==72546014933]["DATE"]
X2_temp = station_data[station_data["STATION"]==72546014933]["TMP"]
X3_date = station_data[station_data["STATION"]==72450003928]["DATE"]
X3_temp = station_data[station_data["STATION"]==72450003928]["TMP"]
```

```python
plt.plot(X1_date,X1_temp,color='k')
plt.plot(X2_date,X2_temp,color='b')
plt.plot(X3_date,X3_temp,color='r')
plt.show()
```

```python
#visualizing the whole dataset using figure and subplot
fig_Station=plt.figure(figsize=(8,13))
ax_Station1=fig_Station.add_subplot(3,1,1)
ax_Station1.plot(X1_date,X1_temp,color='r')
ax_Station2=fig_Station.add_subplot(3,1,2)
ax_Station2.plot(X2_date,X2_temp,color='b')
ax_Station3=fig_Station.add_subplot(3,1,3)
ax_Station3.plot(X3_date,X3_temp,color='g')
plt.xlabel("temp")
plt.ylabel("stationID")
```

```python
station_data["TMP"]=station_data["TMP"].str.replace(",",".").astype(float)
station_data["TMP"]
```

```python
X1_date = pd.to_datetime(station_data[station_data["STATION"]==72356013968]["DATE"],format='%d/%m/%Y %H:%M')
X1_temp = (5/9)*(station_data[station_data["STATION"]==72356013968]["TMP"]-32)
X2_date = pd.to_datetime(station_data[station_data["STATION"]==72546014933]["DATE"],format='%d/%m/%Y %H:%M')
X2_temp = (5/9)*(station_data[station_data["STATION"]==72546014933]["TMP"]-32)
X3_date = pd.to_datetime(station_data[station_data["STATION"]==72450003928]["DATE"],format='%d/%m/%Y %H:%M')
X3_temp = (5/9)*(station_data[station_data["STATION"]==72450003928]["TMP"]-32)
```

```python
X1_date
X1_temp
```

```python
plt.figure()
plt.plot(X1_date,X1_temp,color='r',label="Station1")
plt.plot(X2_date,X2_temp,color='b',label="Station2")
plt.plot(X3_date,X3_temp,color='g',label="Station3")
plt.legend()
plt.xlabel("Date")
plt.ylabel("avg temp")
plt.show()
```

```python
#plot visualization
fig_Station_new=plt.figure(figsize=(8,13))
ax1_Station1=fig_Station_new.add_subplot(3,1,1)
ax1_Station1.plot(X1_date,X1_temp,color='r')
ax1.legend()
ax2_Station2=fig_Station_new.add_subplot(3,1,2)
ax2_Station2.plot(X2_date,X2_temp,color='b')
ax2.legend()
ax3_Station3=fig_Station_new.add_subplot(3,1,3)
ax3_Station3.plot(X3_date,X3_temp,color='g')
ax3.legend()
```

```python
#scatter visualization
fig_Station_new=plt.figure(figsize=(8,13))
ax1_Station1=fig_Station_new.add_subplot(3,1,1)
ax1_Station1.scatter(X1_date,X1_temp,color='r')
ax2_Station2=fig_Station_new.add_subplot(3,1,2)
ax2_Station2.scatter(X2_date,X2_temp,color='b')
ax3_Station3=fig_Station_new.add_subplot(3,1,3)
ax3_Station3.scatter(X3_date,X3_temp,color='g')
```

```python
#create a 2*2 figure to compare scatter and lines
fig=plt.figure(figsize=(15,8))

ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.scatter(X1_date,X1_temp,color='r',label="station 1")
ax1.scatter(X2_date,X2_temp,color='y',label="station 2")
ax1.scatter(X3_date,X3_temp,color='b',label="station 3")
ax1.legend()
ax1.set_xlabel('Date')
ax1.set_ylabel('avg temp')
ax1.set_title("daily avg temp")

ax2.plot(X1_date,X1_temp,color='r',label="station 1")
ax2.plot(X2_date,X2_temp,color='y',label="station 2")
ax2.plot(X3_date,X3_temp,color='b',label="station 3")
ax2.legend()
ax2.set_xlabel('Date')
ax2.set_ylabel('avg temp')
ax2.set_title("daily avg temp")

ax3.scatter(X1_date[0:20],X1_temp[0:20],color='g',label="Station1")
ax3.legend()
ax3.set_xlabel('date')
ax3.set_ylabel('avg temp')

ax4.plot(X3_date[0:20],X3_temp[0:20],color='b',label="Station3")
ax4.legend()
ax4.set_xlabel('date')
ax4.set_ylabel('avg temp')

```

```python
#adding polish
import matplotlib_inline
plt.figure()
plt.plot(X1_date,X1_temp,color="b",label="Station1")
plt.plot(X2_date,X2_temp,color="g",label="Station2")
plt.plot(X3_date,X3_temp,color="r",label="Station3")
ax4.legend()
ax4.set_xlabel('date')
ax4.set_ylabel('avg temp')
plt.title("daily avg temp" , fontsize=14,fontweight='bold' )
plt.grid(linewidth=0.2)
plt.axhline(X1_temp.mean(),color='tab:purple',linestyle="dotted")
plt.axhline(X2_temp.mean(),color='tab:blue',linestyle="dotted")
plt.axhline(X3_temp.mean(),color='tab:orange',linestyle="dotted")
```

```python
#adding polish with alpha
import matplotlib_inline
plt.figure()
plt.plot(X1_date,X1_temp,color="b",label="Station1",alpha=0.3)
plt.plot(X2_date,X2_temp,color="g",label="Station2")
plt.plot(X3_date,X3_temp,color="r",label="Station3",alpha=0.3)
plt.legend()
plt.xlabel('date')
plt.ylabel('avg temp')
plt.title("daily avg temp" , fontsize=14,fontweight='bold' )
plt.grid(linewidth=0.2)
plt.axhline(X1_temp.mean(),color='tab:purple',linestyle="dotted")
plt.axhline(X2_temp.mean(),color='tab:blue',linestyle="dotted")
plt.axhline(X3_temp.mean(),color='tab:orange',linestyle="dotted")
```

```python
#fluctuating temp data 
X1_temp_max = (5/9)*(station_data[station_data["STATION"]==72356013968]["TMP"]+20-32)
X1_temp_min = (5/9)*(station_data[station_data["STATION"]==72356013968]["TMP"]-20-32)
X2_temp_max = (5/9)*(station_data[station_data["STATION"]==72546014933]["TMP"]+20-32)
X2_temp_min = (5/9)*(station_data[station_data["STATION"]==72546014933]["TMP"]-20-32)
X3_temp_max = (5/9)*(station_data[station_data["STATION"]==72450003928]["TMP"]+20-32)
X3_temp_min = (5/9)*(station_data[station_data["STATION"]==72450003928]["TMP"]-20-32)
```

```python
plt.figure()
plt.plot(X1_date,X1_temp,color="b",label="Station1")
plt.plot(X2_date,X2_temp,color="g",label="Station2")
plt.plot(X3_date,X3_temp,color="r",label="Station3")

plt.fill_between(X1_date,X1_temp_max,X1_temp_min,color='b',alpha=0.3)
plt.fill_between(X2_date,X2_temp_max,X2_temp_min,color='g',alpha=0.3)
plt.fill_between(X3_date,X3_temp_max,X3_temp_min,color='r',alpha=0.3)

plt.legend()
plt.xlabel('date')
plt.ylabel('avg temp')
plt.title("daily avg temp" , fontsize=14,fontweight='bold' )
plt.grid(linewidth=0.2)

plt.axhline(X1_temp.mean(),color='b',linestyle="dotted",alpha=0.3)
plt.axhline(X2_temp.mean(),color='g',linestyle="dotted",alpha=0.3)
plt.axhline(X3_temp.mean(),color='r',linestyle="dotted",alpha=0.3)
```

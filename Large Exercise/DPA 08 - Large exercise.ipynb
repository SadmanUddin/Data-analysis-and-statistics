## 8.2 Calculations 

In this section we are going to calculate some aggregate results per year and per decade. Specifically, we are interested in the average and standard deviation of the scores per year/decade. In this way we can hopefully answer the question which year/decade was rated the best for cinema. We are going to do this by writing a function that does these calculations for us

### 8.2.1 Write a function that calculates metrics per year

- Mean of 'avg_vote'
- Standard deviation of 'avg_vote'
- Mean of 'standardized_vote'
- Standard deviation of 'standardized_vote'
- Mean of 'duration'
- Standard deviation of 'duration'
- Total amount of movies

Use this function to answer the following questions:

1. In what year where most movies produced?
2. Which year had the highest average vote?
3. In what year were the standardized votes the least spread out?
4. Which year has the shortest movies?

```python
#overall metrics
merge_Mean_avgVote = merge['avg_vote'].mean()
merge_std_avgVote = merge['avg_vote'].std()
standardize_vote_mean = standardize_vote.mean()
merge_mean_duration = merge['duration'].mean()
merge_std_duration = merge['duration']
total_movies = merge['title'].count()
print(merge_Mean_avgVote)
print(merge_std_avgVote)
print(standardize_vote_mean)
print(merge_mean_duration)
print(merge_std_duration)
print(total_movies)
```

```python
#per year metrics
def per_year(merge):
    mean_avgVote = merge.groupby('year')['avg_vote'].mean()
    std_avgVote = merge.groupby('year')['avg_vote'].std()
    mean_duration = merge.groupby('year')['duration'].mean()
    std_duration = merge.groupby('year')['duration'].std()
    standardize_vote_mean1 = standardize_vote.groupby(merge['year']).mean()
    standardize_vote_std1 = standardize_vote.groupby(merge['year']).std()
    total_movies1 = merge.groupby('year')['title'].count()
    df=pd.concat([mean_avgVote,std_avgVote,mean_duration,std_duration,standardize_vote_mean1,standardize_vote_std1,total_movies1],axis=1)
    df.columns = ["mean_avgVote","std_avgVote","mean_duration","std_duration","standardize_vote_mean","standardize_vote_std","total_movies"]
    return df

df = per_year(merge)

df


```

```python
print(df['total_movies'].idxmax()) #the year with the most produced movies
print(df['mean_avgVote'].idxmax()) #the year with the highest avgvote
print(df['standardize_vote_mean'].idxmin()) #least standardized
print(df['mean_duration'].idxmin())#the year with the shortest movies
```

### 8.2.2 Adapt your function to calculates metrics per decade

Only consider decades up to the 2010's.

Keep the same variables as before. Use the new function to answer the same questions as 8.2.1. Do your answers change?

```python
def per_year(merge):
    decade = (merge['year'] //10) *10
    mean_avgVote = merge.groupby(decade)['avg_vote'].mean()
    std_avgVote = merge.groupby(decade)['avg_vote'].std()
    mean_duration = merge.groupby(decade)['duration'].mean()
    std_duration = merge.groupby(decade)['duration'].std()
    standardize_vote_mean1 = standardize_vote.groupby(decade).mean()
    standardize_vote_std1 = standardize_vote.groupby(decade).std()
    total_movies1 = merge.groupby(decade)['title'].count()
    df_new=pd.concat([mean_avgVote,std_avgVote,mean_duration,std_duration,standardize_vote_mean1,standardize_vote_std1,total_movies1],axis=1)
    df_new.columns = ["mean_avgVote","std_avgVote","mean_duration","std_duration","standardize_vote_mean","standardize_vote_std","total_movies"]
    return df_new

df_decade = per_year(merge)
df_decade
```

```python
print(df_decade['total_movies'].idxmax()) #the year with the most produced movies
print(df_decade['mean_avgVote'].idxmax()) #the year with the highest avgvote
print(df_decade['standardize_vote_mean'].idxmin()) #least standardized
print(df_decade['mean_duration'].idxmin())#the year with the shortest movies
```

### 8.3.4 Calculate the difference in rating between us votes and non-us votes

Create a function to calulate the difference in rating for each movie. Show this yearly difference in a histogram. Show the median and standard deviation.

```python
def per_year_rating_difference(merge):
    Median_us_votes = merge.groupby('year')['us_voters_rating'].median()
    Median_non_us_votes = merge.groupby('year')['non_us_voters_rating'].median()
    std_us_votes = merge.groupby('year')['us_voters_rating'].std()
    std_non_us_votes = merge.groupby('year')['non_us_voters_rating'].std()
    diff_US_nonUs_Median = Median_us_votes-Median_non_us_votes
    diff_US_nonUs_std = std_us_votes-std_non_us_votes
    df_us = pd.concat([diff_US_nonUs_Median,diff_US_nonUs_std],axis=1)
    df_us.columns=['Median_us_nonUS_votes_diff','std_us_nonUs_votes_diff']
    return df_us
df_usNonus1=per_year_rating_difference(merge)
df_usNonus1

plt.hist(df_usNonus1['Median_us_nonUS_votes_diff'],alpha=0.3,bins=20)
plt.hist(df_usNonus1['std_us_nonUs_votes_diff'],alpha=0.3,bins=20)
plt.show()
```

### 8.3.5 Show the difference in rating between us votes and non-us votes on a plot

Show the caluculated difference on a plot. Also show the standard deviation per year.

```python
plt.figure(figsize=(12,6))
plt.plot(df_usNonus1.index,df_usNonus1['Median_us_nonUS_votes_diff'],label='Median rating diff')
plt.plot(df_usNonus1.index,df_usNonus1['std_us_nonUs_votes_diff'],label='std rating diff')
plt.xlabel('Year')
plt.ylabel('Difference')
plt.legend()
plt.show()
```

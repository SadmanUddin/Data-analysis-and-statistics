measure of central tendency
The first numerical metric to take a look at are measures of central tendency...meaning 'where is the central of the data situated'
MEAN..the first and the most straightforward measure of central tendency is the mean aka avg...

```python
mean=np.mean(pokemons["Total"])
mean
```

```python
pokemons["Total"].mean()
```

this way we can always calculate the avg and also the avg total score of all observation is 435...the question is the mean however a good metric of central tendency in our case?...lets check

```python
plt.hist(pokemons["Total"],bins=50,alpha=0.9)
plt.axvline(mean,0,150,color='r')
plt.show()
```

there are diff cases in which the mean is less than optimal.For the presence of outliers the mean is extremely sensitive...and the solution for this is median...another issue is multiple peakedness(a distribution which has more than one peak(mode))

is mean meaningfull for all numeric data?
--the data needs to be continuous for the mean to be meaningfull...if data is discreate(for example a score out of 20), a mean of 16.3 is possible which is meaningless

The Median
the median is teh second measure of central tendency we always calculate when studying a variable..which is calculated fairly easily


Use the Mean when the data is symmetric and has no extreme outliers

Use the mean when:
	•	Data is roughly symmetric
	•	No extreme values (outliers)
	•	You want to use all numerical values

Examples:
	•	Heights of students in one class
	•	IQ scores
	•	Measurement errors in experiments

Why:
The mean uses every value, so it represents the true center if the distribution is balanced.

Example:

Data: 10, 12, 13, 14, 15

Mean
(10+12+13+14+15)/5 = 12.8

Median = 13

Both are close because the data is balanced.

⸻

Use the Median when the data is skewed or has outliers

Use the median when:
	•	Data is skewed
	•	There are extreme values
	•	The dataset mixes very large and small numbers

Examples:
	•	Income or salary data
	•	House prices
	•	Waiting times

Why:
Outliers pull the mean, but the median ignores extreme values and shows the typical value.

Example:

Data: 10, 12, 13, 14, 100

Mean
(10+12+13+14+100)/5 = 29.8

Median = 13

The mean (29.8) is completely misleading because one outlier ruined it.


```python
median=np.median(pokemons["Total"])
median
```

The median is much more robust against outliers...if the mean and median is close to each other then it means there are not much outliers in the distributions...if any the distribution is more or less symmentric

```python
plt.hist(pokemons["Total"],bins=50,alpha=0.9)
plt.axvline(mean,0,150,color='r',label="Mean")
plt.axvline(median,0,150,color='b',label="Median")
plt.legend()
plt.show()
```

indeed the median and mean are close together...so the center of our data is indeed somewhere around 435-450...if these two measure fall more or less together...its a first proof that we have found the center of our dataset

MEASURES OF DISPERSION
Next to getting an idea of where the center of our data is..."if we were to take a random observation.how far away would it lie from the center of the data"?

the firdt measures of dispersion is are the variance and standard deviation...we first calculate for each observation  the distance it lies from the mean...we than simply take the avg of all these distance...	
    •	If numbers are close to the mean → small variance
	•	If numbers are far from the mean → large variance

variance=s^2 = \frac{\sum (x_i - \bar{x})^2}{n-1}
stddev=(s^2)^(0.5)

```python
variance = np.var(pokemons["Total"]);
stddev = np.sqrt(variance)
print(stddev)
```

we conclude that on avg an observation will lie about 120 points away from the mean...it simply shows us how dispread our data is

PERCENTILES
The second measure of dispersion are the percentiles...
A percentile tells you what percentage of data falls below a certain value.

In plain terms: it shows your position in a dataset compared to everyone else.

⸻

Example

If you are in the 80th percentile, it means:
	•	80% of the data is below you
	•	20% is above you

So you’re performing better than 80% of the group.

Example situations:
	•	Test scores
	•	Height statistics
	•	Income distribution


```python
percentile_01 = np.percentile(pokemons["Total"],1)
percentile_99 = np.percentile(pokemons["Total"],99)
median = np.percentile(pokemons["Total"],50)
print(percentile_01)
print(percentile_99)
print(median)
```

From this we can say that 99% of all observations are higher than 200 and 1% of all obervations are lower than 200 ..also we can say that 99% of all observations are lower than 700 where 1% is higher than 700...and 50 50 is 450...means 50% are higher and lower than 450

some few special percentile
q1=25th
q2=50th
q3=75th
interquartile range(IQR) = q3-q1 #measure of dispersion based on the percentiles/quartiles

```python
percentile_q1 = np.percentile(pokemons["Total"],25)
percentile_q2 = np.percentile(pokemons["Total"],50)
percentile_q3 = np.percentile(pokemons["Total"],75)
IQR=percentile_q3-percentile_01
print(percentile_q1)#25th value of total dataset
print(percentile_q2)#50th value of total dataset
print(percentile_q3)#75th value of total dataset
print(IQR)
```

SAMPLE VS POPULATION
to conceptually understand a large number of "Stuff going on" in statistics, it is important to understand the difference between sample and population

population = the entire population of individual you wish to make conclusions for.
sample = the part of the population you have measurements of.

```python
###lets check the effect of the sample size by making a visualisation of some HP stats
size_list = range(1,801)
mean_sample = []
sd_sample = []

for sample_size in size_list:
    mean_sample.append(pokemons["HP"].sample(sample_size).mean())
    sd_sample.append(pokemons["HP"].sample(sample_size).std())
sample_simulation=pd.DataFrame({'sample_size':size_list,'sample_mean':mean_sample,'sample_sd':sd_sample})
sample_simulation.head(10)
```

```python
pokemons["HP"].sample(2)
```

```python
plt.plot(sample_simulation["sample_size"],sample_simulation["sample_mean"],color='b',label='sample avg',alpha=0.7)
plt.plot(sample_simulation["sample_size"],sample_simulation["sample_sd"],color='g',label='sample avg',alpha=0.7)
plt.axhline(pokemons["HP"].mean(),color='r',linestyle="dotted",label = "population avg")
plt.axhline(np.sqrt(np.var(pokemons["HP"])),color='b',linestyle="dotted",label = "population sd")
plt.legend()
plt.show()
```

```python
newdf["Age"].value_counts()
```

```python
age = newdf["Age"].ffill() #this will fill the missing value with the previous last known value...if a position has 1 and after that position is empty or a missing value...ffill(forward fill) will fill the value with 1
age.value_counts()
```

```python
age = newdf["Age"].bfill()#this will fill the missing value with the next known value...if a position has 1 and before that position is empty or a missing value...bfill(backword fill) will fill the value with 1
age.value_counts()
```
